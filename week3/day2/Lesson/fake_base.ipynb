```python
import numpy as np
from faker import Faker
```

```python
fake = Faker()

data = [(fake.name(), fake.email(), fake.passport_gender(), np.random.randint(18, 70), np.random.randint(1000, 10000)) for _ in range(100)]

import pandas as pd
df = pd.DataFrame(data, columns=['Name', 'email', 'gender', 'Age', 'sales'])
print(df.head())
```

```python
import matplotlib.pyplot as plt
plt.bar(df['gender'], df['sales'].mean())
plt.show()
```
