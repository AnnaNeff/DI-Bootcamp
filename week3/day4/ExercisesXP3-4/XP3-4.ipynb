Data visualization is important because it makes numbers easy to see and understand.
It shows patterns, trends, and problems in the data.
It helps people make better decisions and explain information to others.

A line graph is used to show changes over time.
It connects points with lines and makes it easy to see if values go up, down, or stay the same.
It is good for time series, like sales per month or temperature per day.

```python
import matplotlib.pyplot as plt
```

```python

sales = [5000, 5500, 6200, 7000, 7500]
month = ['Jan', 'Feb', 'Maar', 'Apr', 'May']
plt.bar(month, sales)
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.show()
```

```python
import pandas as pd
student = pd.read_csv('Student-Mental-health.csv')
print(student.head())
print(student.info())
```

```python
student['What is your CGPA?'].nunique()
```

```python
import seaborn as sns
import pandas

plt.figure(figsize=(10,4))

viz = sns.histplot(student['What is your CGPA?'], color='purple')
plt.ylabel('Number of students')
plt.xlabel('CGPA')
plt.title('Number of students per CGPA')
plt.gca().set_facecolor("lavender")

for p in viz.patches:
    height = p.get_height()
    if height > 0:
        viz.text(
            p.get_x() + p.get_width()/2.,
            height,
            int(height),
            ha="center", va="bottom",
            fontsize=9, color="black", fontweight="bold"
        )

plt.tight_layout()
plt.show()

plt.show()
```
